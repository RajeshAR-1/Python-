```python
#Label Encoder method
```

```python
from sklearn.preprocessing import LabelEncoder

bdge_typ = ('Arch','Beam','Truss','Cantilever','Tied Arch','Suspension','Cable')
bdge_df = pd.DataFrame(bdge_typ, columns=['Bridge_Types'])

labelencoder = LabelEncoder()

bdge_df['Bridge_Types_Cat'] = labelencoder.fit_transform(bdge_df['Bridge_Types'])
bdge_df
```

```python
#OneHotEncoder Method
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown='ignore')
enc_df = pd.DataFrame(enc.fit_transform(bdge_df[['Bridge_Types_Cat']]).toarray())

bdge_df = bdge_df.join(enc_df)
bdge_df
```
